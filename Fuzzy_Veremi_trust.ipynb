```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

# Split and train
X = df.drop(columns=['attack', 'attack_type', 'Unnamed: 0'])  # drop unwanted columns
y = df['attack']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Get prediction probabilities
probs = model.predict_proba(X_test)
# For binary classification, use the probability of class 1 (attack)
confidence_scores = probs[:, 1]  # class 1 confidence
```

```python
X_test = X_test.copy()  # to avoid warnings
X_test['ml_confidence'] = confidence_scores
```

```python
# Replace this:
# df['ml_confidence'] = np.random.rand(len(df))

# With this:
df_test_with_conf = X_test.copy()
df_test_with_conf['avg_spd_noise'] = (df_test_with_conf['spd_noise_0'] + df_test_with_conf['spd_noise_1']) / 2
df_test_with_conf['avg_acl_noise'] = (df_test_with_conf['acl_noise_0'] + df_test_with_conf['acl_noise_1']) / 2
```

```python
# Define fuzzy variables
spd_noise = ctrl.Antecedent(np.arange(0, 0.05, 0.001), 'spd_noise')
acl_noise = ctrl.Antecedent(np.arange(0, 0.05, 0.001), 'acl_noise')
confidence = ctrl.Antecedent(np.arange(0, 1.01, 0.01), 'confidence')

trust = ctrl.Consequent(np.arange(0, 1.01, 0.01), 'trust')
```

```python
# Speed Noise
spd_noise['low'] = fuzz.trimf(spd_noise.universe, [0, 0, 0.015])
spd_noise['medium'] = fuzz.trimf(spd_noise.universe, [0.01, 0.025, 0.04])
spd_noise['high'] = fuzz.trimf(spd_noise.universe, [0.03, 0.05, 0.05])

# Acceleration Noise
acl_noise['low'] = fuzz.trimf(acl_noise.universe, [0, 0, 0.015])
acl_noise['medium'] = fuzz.trimf(acl_noise.universe, [0.01, 0.025, 0.04])
acl_noise['high'] = fuzz.trimf(acl_noise.universe, [0.03, 0.05, 0.05])

# Model Confidence
confidence['low'] = fuzz.trimf(confidence.universe, [0, 0, 0.4])
confidence['medium'] = fuzz.trimf(confidence.universe, [0.3, 0.5, 0.7])
confidence['high'] = fuzz.trimf(confidence.universe, [0.6, 1.0, 1.0])

# Trust Output
trust['low'] = fuzz.trimf(trust.universe, [0, 0, 0.4])
trust['medium'] = fuzz.trimf(trust.universe, [0.3, 0.5, 0.7])
trust['high'] = fuzz.trimf(trust.universe, [0.6, 1.0, 1.0])
```

```python
rule1 = ctrl.Rule(spd_noise['low'] & acl_noise['low'] & confidence['high'], trust['high'])
rule2 = ctrl.Rule(spd_noise['medium'] | acl_noise['medium'] | confidence['medium'], trust['medium'])
rule3 = ctrl.Rule(spd_noise['high'] & acl_noise['high'] & confidence['low'], trust['low'])
```

```python
trust_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
trust_sim = ctrl.ControlSystemSimulation(trust_ctrl)

# Example on first 5 rows
for i in range(5):
    trust_sim.input['spd_noise'] = df['avg_spd_noise'].iloc[i]
    trust_sim.input['acl_noise'] = df['avg_acl_noise'].iloc[i]
    trust_sim.input['confidence'] = df['ml_confidence'].iloc[i]
    trust_sim.compute()
    print(f"Trust Score for sample {i}: {trust_sim.output['trust']:.2f}")
```

```python
import numpy as np
import pandas as pd
import skfuzzy as fuzz
from skfuzzy import control as ctrl
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Step 1: Load your dataset
df = pd.read_csv('Veremi_final_dataset.csv', nrows=10000)  # Replace with your actual dataset path

# Step 2: Prepare features and labels
X = df[['spd_noise_0', 'acl_noise_0']]  # Select your key features
y = df['attack']  # Target variable

# Step 3: Split data for training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Train a simple model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]  # Confidence score for class "1"

# Step 5: Create new DataFrame with test samples and confidence
df_test_with_conf = X_test.copy()
df_test_with_conf['ml_confidence'] = y_proba

# Step 6: Define fuzzy variables
spd_noise = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'spd_noise')
acl_noise = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'acl_noise')
confidence = ctrl.Antecedent(np.arange(0, 1.1, 0.1), 'confidence')
trust = ctrl.Consequent(np.arange(0, 1.1, 0.1), 'trust')

# Membership functions
spd_noise['low'] = fuzz.trimf(spd_noise.universe, [0, 0, 0.5])
spd_noise['high'] = fuzz.trimf(spd_noise.universe, [0.5, 1, 1])
acl_noise['low'] = fuzz.trimf(acl_noise.universe, [0, 0, 0.5])
acl_noise['high'] = fuzz.trimf(acl_noise.universe, [0.5, 1, 1])
confidence['low'] = fuzz.trimf(confidence.universe, [0, 0, 0.5])
confidence['high'] = fuzz.trimf(confidence.universe, [0.5, 1, 1])
trust['low'] = fuzz.trimf(trust.universe, [0, 0, 0.5])
trust['medium'] = fuzz.trimf(trust.universe, [0.3, 0.5, 0.7])
trust['high'] = fuzz.trimf(trust.universe, [0.5, 1, 1])

# Step 7: Define rules
rule1 = ctrl.Rule(spd_noise['high'] | acl_noise['high'] | confidence['low'], trust['low'])
rule2 = ctrl.Rule(spd_noise['low'] & acl_noise['low'] & confidence['high'], trust['high'])
rule3 = ctrl.Rule(confidence['high'] & (spd_noise['high'] | acl_noise['high']), trust['medium'])

# Step 8: Create control system and simulator
trust_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
trust_sim = ctrl.ControlSystemSimulation(trust_ctrl)

# Step 9: Apply fuzzy system to each sample
trust_scores = []

for i in range(len(df_test_with_conf)):
    try:
        trust_sim.input['spd_noise'] = df_test_with_conf['spd_noise_0'].iloc[i]
        trust_sim.input['acl_noise'] = df_test_with_conf['acl_noise_0'].iloc[i]
        trust_sim.input['confidence'] = df_test_with_conf['ml_confidence'].iloc[i]
        trust_sim.compute()
        trust_scores.append(trust_sim.output['trust'])
    except:
        trust_scores.append(np.nan)  # Fallback for any errors

df_test_with_conf['trust_score'] = trust_scores

# Step 10: Display results
print(df_test_with_conf[['spd_noise_0', 'acl_noise_0', 'ml_confidence', 'trust_score']].head())
```
